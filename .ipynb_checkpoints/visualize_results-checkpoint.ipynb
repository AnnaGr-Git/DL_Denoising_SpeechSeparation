# Plot spectogram of soundfile

```python
# Get soundfiles
path_mix = "data/mix_9026-282386-0021_7809-108638-0035.wav"
path_s1 = "data/s1_9026-282386-0021_7809-108638-0035.wav"
path_s2 = "data/s2_9026-282386-0021_7809-108638-0035.wav"
path_noise = "data/noise_9026-282386-0021_7809-108638-0035.wav"

sound_mix = sf.read(path_mix)[0]
sound_s1 = sf.read(path_s1)[0]
sound_s2 = sf.read(path_s2)[0]
sound_noise = sf.read(path_noise)[0]
```

```python
display(Audio(path_s1))
display(Audio(path_s2))
```

```python
colormap = librosa.display.cmap(librosa.amplitude_to_db(np.abs(librosa.stft(sound_mix))), cmap_seq='GnBu', cmap_bool='gray_r', cmap_div='Blues')

def show_magspec(waveform, **kw):
    return librosa.display.specshow(
        librosa.amplitude_to_db(np.abs(librosa.stft(waveform))),
        cmap=colormap,
        y_axis="log", x_axis="time",
        **kw
    )

f = plt.figure(figsize=(17,3))

img = show_magspec(sound_mix, sr=8000)
plt.xlabel('Time [s]', fontsize=16)
plt.ylabel('Frequency [Hz]', fontsize=16)
cbar = plt.colorbar(img,format="%+2.f dB")
cbar.ax.tick_params(labelsize=16)
plt.title("Mixture", fontsize=19)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

f2 = plt.figure(figsize=(15,3))

plt.subplot(1, 3, 1)
img = show_magspec(sound_s1, sr=8000)
plt.title("S1: Women", fontsize=19)
plt.xlabel('Time [s]', fontsize=16)
plt.ylabel('Frequency [Hz]', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)


plt.subplot(1, 3, 2)
img = show_magspec(sound_s2, sr=8000)
plt.title("S2: Man", fontsize=19)
plt.ylabel('', fontsize=1)
plt.xlabel('Time [s]', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks([])


plt.subplot(1, 3, 3)
img = show_magspec(sound_noise, sr=8000)
plt.title("Noise", fontsize=19)
plt.ylabel('', fontsize=1)
plt.xlabel('Time [s]', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks([])


f.savefig('reports/figures/spectogram_mix.png', bbox_inches="tight",dpi=700)
f2.savefig('reports/figures/spectogram_separate.png', bbox_inches="tight",dpi=700)
```

## Original vs reconstructed signal

```python
# Get soundfiles
path_orig_mix = "data/test2_easy/original/17-362-0027_3835-178030-0019_mix.wav"
path_orig_s1 = "data/test2_easy/original/17-362-0027_3835-178030-0019_s1.wav"
path_orig_s2 = "data/test2_easy/original/17-362-0027_3835-178030-0019_s2.wav"
path_orig_noise = "data/test2_easy/original/17-362-0027_3835-178030-0019_noise.wav"

path_2ch_s1 = "data/test2_easy/reconstructed/svoice/2_channels/17-362-0027_3835-178030-0019_s1.wav"
path_2ch_s2 = "data/test2_easy/reconstructed/svoice/2_channels/17-362-0027_3835-178030-0019_s2.wav"
path_3ch_s1 = "data/test2_easy/reconstructed/svoice/3_channels/17-362-0027_3835-178030-0019_s1.wav"
path_3ch_s2 = "data/test2_easy/reconstructed/svoice/3_channels/17-362-0027_3835-178030-0019_s2.wav"
path_3ch_noise = "data/test2_easy/reconstructed/svoice/3_channels/17-362-0027_3835-178030-0019_s3.wav"

sound_orig_mix, samplerate = sf.read(path_orig_mix)
sound_orig_s1,_ = sf.read(path_orig_s1)
sound_orig_s2,_ = sf.read(path_orig_s2)
sound_orig_noise,_ = sf.read(path_orig_noise)

sound_2ch_s1,_ = sf.read(path_2ch_s1)
sound_2ch_s2,_ = sf.read(path_2ch_s2)
sound_3ch_s1,_ = sf.read(path_3ch_s1)
sound_3ch_s2,_ = sf.read(path_3ch_s2)
sound_3ch_noise,_ = sf.read(path_3ch_noise)
```

```python
display(Audio(path_orig_s1))
display(Audio(path_2ch_s1))
display(Audio(path_3ch_s1))
```

```python
n_samples = len(sound_orig_mix)
t_audio = n_samples/samplerate
times = np.linspace(0, t_audio, num=n_samples)
```

```python
plt.figure(figsize=(15, 5))
plt.plot(times, sound_2ch_s1/max(sound_2ch_s1), label="2channels", linewidth=0.5, alpha=0.5)
plt.plot(times, sound_3ch_s1/max(sound_3ch_s1), label="3channels", linewidth=0.5, alpha=0.5)
plt.plot(times, sound_orig_s1/max(sound_orig_s1), label="original", linewidth=0.5, alpha=0.3)


plt.title('Original vs Reconstructed Signals: S1')
plt.ylabel('Signal Value')
plt.xlabel('Time (s)')
plt.xlim(0, t_audio)
plt.legend()
```

# Plot testing results

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Noise categories

```python
results_with_noise = {'Typing': 4.70,
                      'Nature': 4.83,
                      'Cafe': 4.25,
                      'Babble': 3.50,
                      'Washing\nMachine': 1.91,
                      'Square': 3.6}


results_without_noise = {'Typing': 4.79,
                         'Nature': 4.9,
                         'Cafe': 4.51,
                         'Babble': 2.93,
                         'Washing\nMachine': 2.38,
                         'Square': 3.82}
```

```python
categories = list(results_with_noise.keys())
categories
```

```python
values_with_noise = list(results_with_noise.values())
values_without_noise = list(results_without_noise.values())
```

```python
X_axis = np.arange(len(categories))
fig, ax = plt.subplots(1, figsize=(15,7))
ax.bar(x=X_axis-0.2, height=values_without_noise, width=0.3, label = 'SVoice 2ch', color='#BCB9FB', edgecolor='black', alpha = 1)
ax.bar(x=X_axis+0.2, height=values_with_noise, width=0.3, label = 'SVoice 2+1ch', color='#8FCBED', edgecolor='black', alpha = 1)

ax.set_xticks(X_axis, categories, fontsize=27)
plt.yticks(fontsize=27)

plt.xlabel("Noise Categories", fontsize=27)
plt.ylabel("SI-SNRi [dB]", fontsize=27)
plt.title("Speech Separation Performance\nwith Different Noise Categories", fontsize=32)
plt.legend(fontsize=25)

fig.savefig('reports/figures/noise_categories.png', bbox_inches="tight",dpi=700)
```

### Noise intensities

```python
intensities_with_noise = {'-3': 3.17,
                          '0': 4.25,
                          '3': 4.68,
                          '5': 4.78,
                          'No noise': 6.41}


intensities_without_noise = {'-3': 3.72,
                             '0': 4.51,
                             '3': 4.62,
                             '5': 4.62,
                             'No noise': 6.3}
```

```python
intensities = list(intensities_with_noise.keys())
values_intens_with_noise = list(intensities_with_noise.values())
values_intens_without_noise = list(intensities_without_noise.values())
```

```python
fig, ax = plt.subplots(1, figsize=(15,7))
ax.plot(intensities, values_intens_without_noise, label='SVoice 2ch', color='#BCB9FB', linewidth=4)
ax.plot(intensities, values_intens_with_noise, label='SVoice 2+1ch', color='#8FCBED', linewidth=4)
plt.xlabel("Input SNR (Noise Intensity)", fontsize=27)
plt.ylabel("SI-SNRi [dB]", fontsize=27)
plt.legend(fontsize=27)
plt.xticks(fontsize=27)
plt.yticks(fontsize=27)

plt.title('Speech Separation Performance in Caf-Scenario\nwith Different Noise Intensities', fontsize=32)

fig.savefig('reports/figures/noise_intensities.png', bbox_inches="tight",dpi=700)
```

```python
fig, ax = plt.subplots(figsize =(25, 15))
boxprops = dict(linestyle='-', linewidth=3, color='k')
medianprops = dict(linestyle='-', linewidth=2.5, color='k')

columns = [svoice_2s, svoice_3s, DPRNN_2s, DPRNN_3s, svoice_first_denoise]
bplot = ax.boxplot(columns, notch=False, patch_artist=True, meanline=True, showmeans=False, widths=0.8, boxprops=boxprops, medianprops=medianprops)

colors = ['#BCB9FB', '#8FCBED', '#ACE0F4', '#B1CDF9', '#546FB6']
for patch, color in zip(bplot['boxes'], colors):
      patch.set_facecolor(color)

plt.xticks([1, 2, 3, 4,5], ["SVoice 2ch", "SVoice 2+1ch", "DPRNN 2ch", "DPRNN 2+1ch", "SVoice cascade"], fontsize=34)
plt.yticks(fontsize=34)
#plt.title("Speech Separation Performance on noisy LibriMix Data", fontsize=38)
plt.ylabel('SI-SNRi [dB]', fontsize=34)
fig.savefig('reports/figures/boxplot.png', bbox_inches="tight",dpi=700)
plt.show()
```
